위와 같이 UNK가 필요한 경우에 아래와 같은 해결책을 사용할 수도 있다.

```python
code = """
from gensim.models import Word2Vec
model = Word2Vec.load('/gdrive/MyDrive/colab/week10/ko.bin')
import numpy as np
rand_vector = np.random.randn(200)
model.wv.add('unk',rand_vector)
model.build_vocab([['unk']], update=True)
print(model.wv.most_similar('unk'))
"""

!python3.7 -c "$code"
```

```python
# 마지막 코드는 부득이하게도 에러가 납니다.
code = """
from gensim.models import Word2Vec
model = Word2Vec.load('/gdrive/MyDrive/colab/week10/ko.bin')
positive_list = ['컴퓨터공학', '인문학']
negative_list = ['자연과학']
for i in range(len(positive_list) + len(negative_list)):
  try:
    print(model.wv.most_similar(positive=positive_list, negative=negative_list)[:3])
  except Exception as e:
    error = str(e)
    error_word = error.split(' ')[1].strip('\'')
    for idx, word in enumerate(positive_list):
      if word == error_word:
        positive_list[idx] = 'unk'
        print('positive : {}'.format(positive_list))
    for idx, word in enumerate(negative_list):
      if word == error_word:
        negative_list[idx] = 'unk'
        print('negative : {}'.format(negative_list))
"""
!python3.7 -c "$code"
```
