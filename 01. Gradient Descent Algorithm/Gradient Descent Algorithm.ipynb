```python
import numpy as np
import torch

# 데이터 읽기 함수
def load_dataset(file):

  data = np.loadtxt(file)
  print("DATA=",data)

  x = data[:,0:-1]#모든 행 1열
  print("X=",x)

  y = np.reshape(data[:,-1],(3,1))#모든 행 끝 열
  print("Y=",y)

  # 텐서로 변경
  # numpy를 고차원 매트릭스인 tensor로 캐스팅하기 위함
  x = torch.tensor(x, dtype = torch.float)
  y = torch.tensor(y, dtype = torch.float)
  return (x, y)

# 텐서를 리스트로 바꾸는 함수
def tensor2list(input_tensor):
  return input_tensor.cpu().detach().numpy().tolist()

x, y = load_dataset("/gdrive/MyDrive/Colab/Week02/train.txt")

# 가중치 초기화
#랜덤으로 가중치를 1차원짜리 w 결정 1*1 행렬을 가우시안 분포로서 랜덤값을 갖고 와서 3으로 늘리기
#(3 곱하는 건 사실 큰 의미는 없음)
w = torch.randn(1)*3.0
#텐서를 리스트로 바꾼 뒤에 찍기
print("\nInit: w = {0}\n".format(tensor2list(w)))

# 학습률
alpha = 0.1

# 학습데이터 수
#y가 3행 1열이니까 1열의 행의 수로 학습 데이터 수 도출
m = y.size(0)

for epoch in range(50):
  #epoch은 학습 데이터를 전부 다 읽었을 때를 1번의 epoch이라고 표현. 일종의 학습 횟수라고 볼 수 있음
  #batch는 gradient를 계산하는 단위로 전체 데이터를 조금 분할하는데 그 분할된 단위를 batch라고 한다. w를 업데이트할 때는 batch 단위
  #batch에서는 epoch에서 데이터를 뽑을 때 랜덤으로 뽑아서 여러 번 학습하는데 이때 어떤 데이터는 뽑히지 않을 수 있으나
  #이는 데이터의 순서에 의한 학습을 방지하기 위함이다.
  #batch의 필요성은 데이터가 굉장히 클 때 메모리를 효율적으로 운용하기 위해서 쪼개서 사용
  #여기서는 epoch이 작아서 그냥 한 것
  cost = 0
  for i in range(m):
    cost = cost + ((w*x[i])-y[i])*x[i]
  cost = cost * 1/m
  #여기서는 m을 batch로 취급하고 batch 한 번 수행한 것을 1step이라고 함
  w = w - alpha * cost
  #이렇게 cost function 연산 완성

  print("Epoch {0:d}: w = {1}".format(epoch+1,tensor2list(w)))
```
